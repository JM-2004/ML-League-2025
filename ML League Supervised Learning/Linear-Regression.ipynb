## 4. Data Preprocessing

### 🧩 4.1 Imputing the `missing values`

Missing data is common in real-world datasets and needs to be handled before training a machine learning model.


#### ❓ Why Handle Missing Values?

* Algorithms like **Linear Regression**, **SVM**, etc., **cannot handle NaNs**.
* Missing values can bias results or reduce model performance.
* Ensures the dataset is complete for training and inference.


#### ✅ Best Practices

* Impute **after splitting** train/test if you're using statistics from training data.
* Track which values were imputed (e.g., using an extra binary column).
* Use domain knowledge to guide imputation choices.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean' , missing_values=np.nan)
```

```python
imputer.fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
test_inputs[numeric_cols]= imputer.transform(test_inputs[numeric_cols])
```

## 5. Introduction to Linear Regression

### 🔹 What is Linear Regression?

Linear Regression is a **supervised learning algorithm** used to model the relationship between a **dependent variable (target)** and one or more **independent variables (features)** by fitting a linear equation to observed data.

### 🔸 Simple Linear Regression

* One independent variable.
* Equation:

  $$
  y = \beta_0 + \beta_1 x + \epsilon
  $$

### 🔸 Multiple Linear Regression

* Multiple independent variables.
* Equation:

  $$
  y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon
  $$

### Objective

* Minimize the **difference** between predicted values ($\hat{y}$) and actual values ($y$).
* Done by **minimizing a cost function**, typically the **Mean Squared Error (MSE)**.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Initialize the model
model = LinearRegression()

# Train the model on training data
model.fit(X_train, train_targets)

# Predict on validation set
val_preds = model.predict(X_val)

# Predict on test set
test_preds = model.predict(X_test)
```

## 🛠️ Evaluation Metrics

### 1. **Mean Absolute Error (MAE)**

* Average of absolute errors.
* Formula:

  $$
  MAE = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
  $$
* Easy to understand, not sensitive to outliers.


### 2. **Mean Squared Error (MSE)**

* Average of squared differences between predicted and actual values.
* Formula:

  $$
  MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
  $$
* Penalizes larger errors more than MAE.


### 3. **Root Mean Squared Error (RMSE)**

* Square root of MSE.
* Formula:

  $$
  RMSE = \sqrt{MSE}
  $$
* Same units as target variable, interpretable.


### 4. **R-squared (R² Score)**

* Proportion of variance in dependent variable that is predictable from the independent variables.
* Formula:

  $$
  R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
  $$
* Range: \[0, 1] (closer to 1 is better).
* $R^2 = 0$: Model explains none of the variability.
* $R^2 = 1$: Perfect fit.

```python
# Evaluate on validation set
print("Validation Metrics:")
print("MAE:", mean_absolute_error(val_targets, val_preds))
print("RMSE:", np.sqrt(mean_squared_error(val_targets, val_preds)))
```

## Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Initialize the Decision Tree model with hyperparameters
model = DecisionTreeRegressor(
    max_depth=10,               # limits depth of the tree (prevents overfitting)
    min_samples_split=5,        # minimum samples required to split an internal node
    min_samples_leaf=2,         # minimum samples required at a leaf node
    random_state=42             # for reproducibility
)

# Train the model on training data
model.fit(X_train, train_targets)

# Predict on validation set
val_preds = model.predict(X_val)

# Predict on test set
test_preds = model.predict(X_test)
```

```python
# Evaluate on validation set
print("Validation Metrics:")
print("MAE:", mean_absolute_error(val_targets, val_preds))
print("RMSE:", np.sqrt(mean_squared_error(val_targets, val_preds)))
```
